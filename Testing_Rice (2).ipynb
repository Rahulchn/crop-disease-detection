```python
import numpy as np
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
import os

cnn = tf.keras.models.load_model("C:\\Users\\Rahul Chauhan\\OneDrive\\Desktop\\ALL\\Project_Data_Rice\\trained_rice_leaf_model.keras")

image_path = "C:\\Users\\Rahul Chauhan\\OneDrive\\Desktop\\ALL\\Project Dataset Rice\\Rice_Leaf_Diease\\Rice_Leaf_Diease\\test\\healthy\\healthy (341).jpg"
image_name = os.path.basename(image_path) 
image = tf.keras.preprocessing.image.load_img(image_path, target_size=(128, 128))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.expand_dims(input_arr, axis=0)

predictions = cnn.predict(input_arr)
result_index = np.argmax(predictions)

class_names = [
    "Bacterial Leaf Blight", "Brown Spot", "Healthy", "Leaf Blast", "Leaf Scald",
    "Narrow Brown Spot", "Neck Blast", "Rice Hispa", "Sheath Blight", "Tungro"
]

predicted_class = class_names[result_index]

img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6, 6))
plt.imshow(img)
plt.title(f"Prediction: {predicted_class}", fontsize=12, fontweight='bold')
plt.axis("off")
plt.show()

print(f"Image Name: {image_name}")
print(f"Predicted Disease: {predicted_class}")
```
